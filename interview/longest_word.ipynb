```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.appName("test").getOrCreate()
```

```python
rdd = spark.sparkContext.textFile("words.txt")
word_lengths = rdd.map(lambda word: (word, len(word))).sortBy(lambda x: x[1], ascending=False)
longest_word = word_lengths.first()[0]
longest_word
```

```python
df = spark.read.text("words.txt")
df = df.withColumn("length", length(col("value"))).orderBy(col("length").desc())
logest_word = df.collect()[0]['value']
logest_word
```
